```python
import pandas as pd

def normalize_nested_posts(df, column_name="posts"):
    """
    Normalize a DataFrame containing a nested dictionary in a specific column.

    Args:
        df (pd.DataFrame): Original DataFrame with nested column.
        column_name (str): The column name containing nested data (default is 'posts').

    Returns:
        pd.DataFrame: Flattened DataFrame with expanded nested data.
    """
    if column_name in df.columns:
        # Normalize the nested column
        expanded_df = pd.json_normalize(df[column_name])
        # Concatenate expanded columns back with the original DataFrame (drop the nested column)
        df = pd.concat([df.drop(columns=[column_name]), expanded_df], axis=1)
    return df

# Normalize the DataFrames
viewed_normalized_df = normalize_nested_posts(viewed_df)
liked_normalized_df = normalize_nested_posts(liked_df)
posts_normalized_df = normalize_nested_posts(posts_df)

# Display first few rows of each normalized DataFrame
print("\nViewed Posts Normalized DataFrame:")
print(viewed_normalized_df.head())

print("\nLiked Posts Normalized DataFrame:")
print(liked_normalized_df.head())

print("\nPosts Normalized DataFrame:")
print(posts_normalized_df.head())
```

# Step 2: Data Preparation
We’ll use the all_posts_df dataset to extract metadata for content-based recommendations.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Assuming all_posts_df is already loaded
# Sample data structure
all_posts_df = pd.DataFrame({
    'post_id': [1, 2, 3, 4],
    'category': ['Sports', 'Music', 'Travel', 'Music'],
    'tags': ['football, soccer', 'pop, rock', 'adventure, hiking', 'jazz, classical'],
    'likes': [100, 150, 200, 50]
})

# Combine relevant features into a single text column
all_posts_df['metadata'] = all_posts_df['category'] + " " + all_posts_df['tags']

# Display the processed dataframe
print(all_posts_df)
```

# Step 3: Create TF-IDF Features for Content-Based Recommendation
This step converts text-based metadata into numerical embeddings.

```python
# Convert metadata into TF-IDF features
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(all_posts_df['metadata'])

# Store TF-IDF matrix for similarity computation
all_posts_df['embedding'] = list(tfidf_matrix.toarray())
```

# Step 4: Content-Based Recommendation Function
The function will recommend videos similar to a given video based on metadata.

```python
def recommend_videos(post_id, all_posts, tfidf_matrix, top_n=5):
    try:
        # Get the index of the video with the given post_id
        post_idx = all_posts[all_posts['post_id'] == post_id].index[0]

        # Compute cosine similarity of this video with all other videos
        cosine_similarities = cosine_similarity(tfidf_matrix[post_idx], tfidf_matrix).flatten()

        # Get indices of top_n similar videos
        similar_indices = cosine_similarities.argsort()[-top_n - 1:-1][::-1]

        # Return the recommended videos
        return all_posts.iloc[similar_indices][['post_id', 'category', 'tags']]
    except IndexError:
        print(f"Post ID {post_id} not found in the dataset.")
        return pd.DataFrame()

# Example usage
print("Recommendations for post ID 1:")
recommendations = recommend_videos(1, all_posts_df, tfidf_matrix)
print(recommendations)
```

# Step 5: Trending Recommendations
Identify the most popular videos based on metrics like likes or ratings.

```python
def get_trending_videos(all_posts, top_n=5):
    # Sort videos by likes in descending order
    trending = all_posts.sort_values(by='likes', ascending=False)
    return trending.head(top_n)[['post_id', 'category', 'tags', 'likes']]

# Example usage
print("Trending videos:")
trending_videos = get_trending_videos(all_posts_df)
print(trending_videos)
```

# Step 6: Hybrid Recommendation
Combine content-based and trending recommendations to create a balanced system.

```python
def hybrid_recommendation(post_id, all_posts, tfidf_matrix, top_n=5, trending_weight=0.5):
    # Get content-based recommendations
    content_recs = recommend_videos(post_id, all_posts, tfidf_matrix, top_n=top_n)

    # Get trending recommendations
    trending_recs = get_trending_videos(all_posts, top_n=top_n)

    # Combine and weigh recommendations
    combined = pd.concat([content_recs, trending_recs]).drop_duplicates()
    combined['score'] = combined.index.map(lambda x: trending_weight if x in trending_recs.index else (1 - trending_weight))

    # Sort combined recommendations by score
    return combined.sort_values(by='score', ascending=False).head(top_n)

# Example usage
print("Hybrid recommendations for post ID 1:")
hybrid_recs = hybrid_recommendation(1, all_posts_df, tfidf_matrix)
print(hybrid_recs)
```

```python
def precision_at_k(recommended, actual, k):
    recommended = recommended[:k]
    return len(set(recommended) & set(actual)) / k

# Example usage
recommended_posts = [1, 2, 3, 4]  # Replace with IDs of recommended videos
actual_liked_posts = [2, 3, 5]   # Replace with IDs of actual liked videos
print(f"Precision@3: {precision_at_k(recommended_posts, actual_liked_posts, k=3)}")
```

```python
def click_through_rate(recommended, actual_liked_posts):
    """
    Calculate the Click-Through Rate (CTR)
    """
    # Find how many recommended posts were actually liked
    relevant_recommendations = set(recommended) & set(actual_liked_posts)

    # Calculate CTR
    ctr = len(relevant_recommendations) / len(recommended) if len(recommended) > 0 else 0
    return ctr

# Example usage
recommended_posts = [1, 2, 3, 4, 5]  # Example recommended posts
actual_liked_posts = [2, 3, 5]  # Actual posts that the user clicked on or liked

ctr = click_through_rate(recommended_posts, actual_liked_posts)
print(f"Click-Through Rate (CTR): {ctr:.2f}")
```

```python
def mean_average_precision(recommended_posts_list, actual_liked_posts_list, k=5):
    """
    Calculate Mean Average Precision (MAP) for multiple users
    """
    total_ap = 0
    for recommended, actual in zip(recommended_posts_list, actual_liked_posts_list):
        ap = 0
        relevant_count = 0
        for i in range(min(k, len(recommended))):
            if recommended[i] in actual:
                relevant_count += 1
                ap += relevant_count / (i + 1)

        # Average precision for this query
        if relevant_count > 0:
            ap /= relevant_count
        total_ap += ap

    # Return Mean Average Precision (MAP)
    map_score = total_ap / len(recommended_posts_list) if len(recommended_posts_list) > 0 else 0
    return map_score

# Example usage
recommended_posts_list = [
    [1, 2, 3, 4, 5],  # Recommended posts for user 1
    [2, 3, 5, 1, 6],  # Recommended posts for user 2
    [4, 1, 6, 2, 5]   # Recommended posts for user 3
]

actual_liked_posts_list = [
    [2, 3, 5],  # Actual liked posts for user 1
    [2, 3],     # Actual liked posts for user 2
    [4, 5]      # Actual liked posts for user 3
]

map_score = mean_average_precision(recommended_posts_list, actual_liked_posts_list, k=5)
print(f"Mean Average Precision (MAP): {map_score:.2f}")
```
